```python
#@title
#For training the K-means model  
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x)
```

```python
#@title
#clusters  
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'magenta', label = 'Cluster 1') # 1.cluster  
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'Yellow', label = 'Cluster 2') # 2.cluster  
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'green', label = 'Cluster 3') # 3.cluster  
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'blue', label = 'Cluster 4') # 4.cluster  
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'pink', label = 'Cluster 5') # 5.cluster  
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'Red', label = 'Centroid')   
plt.title('Clusters of the customers')  
plt.xlabel('annual_income')  
plt.ylabel('spending_scores')  
plt.legend()  
plt.show()
```
